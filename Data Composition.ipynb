# Imputing the missing values

To impute the missing values we use 3 techniques most commonly
1. Mean
2. Median
3. Mode

```python
df.isnull().sum()
```

```python
# imputing missing values of age using mean
# firstly we find mean, then median and lastly mode of age column
df.age.mean()
```

```python
# this is median of age
df.age.median()
```

```python
# This is mode of age column 
df.age.mode(1)
```

```python
# now its upto us to use mode mean or median
# commonly median and mean is used for numerical values while mode is used for categorical values
df.age.fillna(df.age.mean(),inplace=True)# we have changed the null or nan values to mean of that column
```

```python
df.isnull().sum()
```

```python
# now you can see that emabrked town and emarkedd are both categorical values
# for categorical values we use mode
df.embark_town.fillna(df.embark_town.mode()[0],inplace=True)
```

```python
df.embarked.fillna(df.embarked.mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
# now we will confirm that there are no other missing values using heatmap
sns.heatmap(df.isnull(),cbar=False)#the heatmap is clear it means there are no kind of values which are missing
```

# Binning

#### Introduction:
Binning is making categories from numerical values to make sense out of it.


```python
# firstly we will start with age group
# we find min, max of age
print("Minimum age ",df.age.min())
print("Maximum age ",df.age.max())
```

```python
# so we know that there can be categories of age like babies, children, teenagers, adults, yound etc for that we use binning
# to find the distribution of age we use histogram graph
sns.histplot(df.age)
```

```python
# Firstly we define binnes and labels for the grouping of data 
binned = [0,1,5,12,18,30,50,80]
labels = ['Babies','Toddlers','Children','Teens','Young','Middle','Old']
df['Binned_data']=pd.cut(df.age,bins=binned,labels=labels)
```

```python
df
```

Renaming a Column

```python
df.rename(columns={'Binned_data':'Age_Group'},inplace=True)
```
